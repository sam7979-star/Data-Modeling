```python
last_load_date = '1900-01-01'
```

###Creating Dynamic Last_load_date if data is already loaded into the table and taking last loaded date from loaded date###

```python
if spark.catalog.tableExists('datamodeling.bronze.bronze_table'):
    last_load_date = spark.sql("select max(order_date) from datamodeling.bronze.bronze_table").collect()[0][0]
                                                                                                #first row of the date value
else:
    last_load_date = '1900-01-01'
```

```python
last_load_date
```

###Selecting Only Incremental Data###

```python
spark.sql(f"""select * from datamodeling.default.source_data where order_date >'{last_load_date}'""").createOrReplaceTempView("bronze_src")
```

```python
%sql
select * from bronze_src
```

```python
%sql
create or replace table datamodeling.bronze.bronze_table
as
select * from bronze_src
```
